# Feature Engineering

## Missing Values Handling For Categorical Variables

```python
for feature in Categorical_features:
    data=house_data.copy()
    if data[feature].isnull().sum()>1 and data[feature].dtypes=='O':
        print(feature+'  has', np.round(data[feature].isnull().mean(),4),'% missing values')
```

**So, these categorical features have missing values.Now our next step is to replace nan values by a new label such as 'Missing'.** 

**In this way we handle missing values in categorical features.**

```python
for feature in Categorical_features:
    #data=house_data.copy()
    if data[feature].isnull().sum()>1 and data[feature].dtypes=='O':
        house_data[feature]=house_data[feature].fillna('Missing')
        print(feature+' has', np.round(house_data[feature].isnull().mean(),4),'% missing values')
```

```python
house_data.head()
```

## Missing Value Handling for Numerical Features 

```python
for feature in numerical_columns_number:
    data=house_data.copy()
    if data[feature].dtype!='O' and data[feature].isnull().sum()>1:
        print(feature+'  has', np.round(data[feature].isnull().mean(),4),'% missing values')
```

```python
feature_with_nan=[feature for feature in numerical_columns_number if house_data[feature].dtype!='O' and house_data[feature].isnull().sum()>1]
feature_with_nan
```

**So, we have only three numerical features which has missing values.**

**Replace all nan values by median or mode value of the corresponding feature. We should not use mean as mean is not robust to outliers.**

```python
for feature in numerical_columns_number:
    #data=house_data.copy()
    if house_data[feature].dtype!='O' and house_data[feature].isnull().sum()>1:
        median_value=house_data[feature].median()
        house_data[feature+'_nan']=np.where(house_data[feature].isnull(),1,0)
        house_data[feature]=house_data[feature].fillna(median_value)
        
```

```python
house_data.head(50)
```

```python
for feature in feature_with_nan:
    #data=house_data.copy()
    if house_data[feature].dtype!='O':
        print(feature+'  has', np.round(house_data[feature].isnull().mean(),4),'% missing values')
```

```python
for feature in Categorical_features:
    temp=house_data.groupby(feature)['SalePrice'].count()/len(house_data)
    temp_df=temp[temp>0.01].index
    house_data[feature]=np.where(house_data[feature].isin(temp_df),house_data[feature],'Rare_VAriable')
```

```python
house_data.head(100)
```

## Convert all categorical variables into numerical variables

```python
for feature in Categorical_features:
    labels_ordered=house_data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    house_data[feature]=house_data[feature].map(labels_ordered)
```
